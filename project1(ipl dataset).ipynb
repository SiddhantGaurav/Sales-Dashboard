```python
batting_first=ipl[ipl["win_by_runs"]!=0]
```

```python
batting_first.head()
```

```python
plt.figure(figsize=(8,8))

plt.bar(list(batting_first["winner"].value_counts()[0:3].keys()),list(batting_first["winner"].value_counts()[0:3]),color=["blue","yellow","green"])
plt.show()
```

```python
plt.figure(figsize=(8,8))

plt.pie(list(batting_first["winner"].value_counts()[0:5]),labels=list(batting_first["winner"].value_counts()[0:5].keys()),autopct="%0.1f%%")
plt.show()
```
